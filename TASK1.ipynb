```python
import seaborn as sns

df = sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt

species_counts = df['species'].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=species_counts.index, y=species_counts.values, palette='viridis')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Bar Chart of Iris Species')
plt.show()
```

```python
plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
sns.histplot(df['sepal_length'], bins=20, kde=True, color='blue')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(df['sepal_width'], bins=20, kde=True, color='green')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(df['petal_length'], bins=20, kde=True, color='red')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(df['petal_width'], bins=20, kde=True, color='purple')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')

plt.tight_layout()
plt.show()
```
